```python
#dropping the unwanted columns from both datasets
```

```python
train.drop(columns=["educationno","race","relationship","sex","capitalgain","capitalloss"],axis=1,inplace=True)
```

```python
test.drop(columns=["educationno","race","relationship","sex","capitalgain","capitalloss"],axis=1,inplace=True)
```

```python
#encoding
from sklearn.preprocessing import LabelEncoder
from sklearn.preprocessing import OneHotEncoder
ohe=OneHotEncoder()
le= LabelEncoder()
```

```python
train["Salary"]=train["Salary"].str.strip()
train["Salary"]=train["Salary"].apply(lambda x: "less" if x=="<=50K" else "high")
```

```python
train
```

```python
test["Salary"]=test["Salary"].str.strip()
test["Salary"]=test["Salary"].apply(lambda x: "less" if x=="<=50K" else "high")
```

```python
#encoding the data using label encoder
```

```python
train["workclass"]=le.fit_transform(train["workclass"])
train["education"]=le.fit_transform(train["education"])
train["maritalstatus"]=le.fit_transform(train["maritalstatus"])
train["occupation"]=le.fit_transform(train["occupation"])
train["native"]=le.fit_transform(train["native"])
train["Salary"]=le.fit_transform(train["Salary"])
```

```python
test["workclass"]=le.fit_transform(test["workclass"])
test["education"]=le.fit_transform(test["education"])
test["maritalstatus"]=le.fit_transform(test["maritalstatus"])
test["occupation"]=le.fit_transform(test["occupation"])
test["native"]=le.fit_transform(test["native"])
test["Salary"]=le.fit_transform(test["Salary"])
```

```python
train
```

```python
test
```

```python
#splitting the data
```

```python
xtrain=train.iloc[:,:-1]
ytrain=train.iloc[:,-1]
```

```python
xtest=test.iloc[:,:-1]
ytest=test.iloc[:,-1]
```

```python
#importing the model
from sklearn.naive_bayes import MultinomialNB
mn=MultinomialNB()
```

```python
mn.fit(xtrain,ytrain)
```

```python
#predicting using xtest
```

```python
ypred= mn.predict(xtest)
```

```python
ypred
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(ytest,ypred))
```

```python
#using another model
```

```python
from sklearn.naive_bayes import CategoricalNB
```

```python
#initialization
cat= CategoricalNB()
```

```python
ypred1= cat.predict(xtest)
```
